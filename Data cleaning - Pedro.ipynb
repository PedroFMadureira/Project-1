```python
# Split the 'height' column into feet and inches
feet_inches = Fifa['height'].str.split("'", expand=True)
feet = feet_inches[0].astype(int)
inches = feet_inches[1].str.strip('"').astype(int)

# Calculate the total inches
total_inches = feet * 12 + inches

# Print the result
Fifa['height'] = total_inches
```

```python
object_columns = Fifa.columns[Fifa.dtypes == 'object']
Fifa[object_columns] = Fifa[object_columns].astype(str)
for col in object_columns:
    Fifa[col] = Fifa[col].str.replace(r'\+.+', '', regex=True)
```
